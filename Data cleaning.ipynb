```python
import pandas as pd

df = pd.read_csv(r"Superstore Sales .csv")
```

```python
df.isnull().sum()
```

```python
duplicates_count = df.duplicated().sum()
print(f"{duplicates_count}")
print("---------")
dup = "Row ID"
duplicates_count = df.duplicated(subset=[dup]).sum()
print(f"'{dup}': {duplicates_count}")
print("---------")
if duplicates_count > 0:
    print(df[df.duplicated(subset=[dup], keep=False)].sort_values(by=dup))
else:
    print("NO dup")
```

```python
num1 = df[df['Row ID'] == 3406]
num2 = df[df['Row ID'] == 3407]

print(num1)
print('-------------------------------------------------------')
print(num2)
```

```python
df = df[df['Row ID'] != 3407]
```
